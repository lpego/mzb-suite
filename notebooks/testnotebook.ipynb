```python
class ToArguments(object):
    def __init__(self, args):
        for key in args:
            setattr(self, key, args[key])

    def __str__(self):
        return self.__dict__.__str__()
```

```python
ROOT_DIR = Path("/data/shared/mzb-workflow/")
MODEL = "efficientnet-b2-v0"

arguments = {
    "config_file": ROOT_DIR /  "configs/configuration_flume_datasets.yaml",
    "input_dir": ROOT_DIR / "data/learning_sets/project_portable_flume/aggregated_learning_sets/val_set/",
    "input_model": ROOT_DIR /f"models/mzb-classification-models/{MODEL}", 
    "output_dir": ROOT_DIR / "results/classification/project_portable_flume/test_",
}
    
with open(str(arguments["config_file"]), "r") as f:
    cfg = yaml.load(f, Loader=yaml.FullLoader)

cfg["trcl_gpu_ids"] = None
cfg
```

```python
# Transforms configurations dicts to argparse arguments
arguments = ToArguments(arguments)
cfg = ToArguments(cfg)
```

```python
inference_classifier(arguments, cfg)
```
